```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('creditcard.csv')

# Display the first few rows of the dataset
df.head()
```

```python
# Check for missing values and data types
df.info()

# Summary statistics
df.describe()
```

```python
# Drop rows with missing values (if any)
df.dropna(inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Count the occurrences of each class (0 = Non-fraud, 1 = Fraud)
sns.countplot(x='Class', data=df)
plt.title('Class Distribution')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Count the occurrences of each class (0 = Non-fraud, 1 = Fraud)
sns.countplot(x='Class', data=df)  # Now, 'data' is defined
plt.title('Class Distribution')
plt.xlabel('Class (0: Non-Fraud, 1: Fraud)')
plt.ylabel('Number of Transactions')
plt.show()
```

```python
# Check for missing values
df.isnull().sum()

# Drop or impute missing values if necessary
df = df.dropna()  # Or use other imputation techniques
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Assuming you have a 'features' DataFrame with your feature columns
features = df.drop(columns=['Class'])  # Drop the target column
target = df['Class']  # Save the target column

features_scaled = scaler.fit_transform(features)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features_scaled, target, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

model1 = LogisticRegression()
model1.fit(X_train, y_train)

# Predictions
y_pred1 = model1.predict(X_test)

# Evaluation
print(confusion_matrix(y_test, y_pred1))
print(classification_report(y_test, y_pred1))
```

```python
from sklearn.ensemble import RandomForestClassifier

model2 = RandomForestClassifier(n_estimators=100)
model2.fit(X_train, y_train)

# Predictions
y_pred2 = model2.predict(X_test)

# Evaluation
print(confusion_matrix(y_test, y_pred2))
print(classification_report(y_test, y_pred2))
```

```python
import pickle

# Save Logistic Regression model
with open('logistic_model.pkl', 'wb') as file:
    pickle.dump(model1, file)

# Save Random Forest model
with open('random_forest_model.pkl', 'wb') as file:
    pickle.dump(model2, file)
```
